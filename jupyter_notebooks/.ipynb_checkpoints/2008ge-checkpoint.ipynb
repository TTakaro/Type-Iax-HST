```python
# Removes points with nan. It is assumed that these are all magnitude measurements, so they are set to their max.
flat = catalog.flatten()
for i in range(flat.size):
    if (np.isnan(flat[i])) or (not np.isfinite(flat[i])):
        flat[i] = 99.999
        catalog[i//271][i%271] = 99.999
        print(i)
```

```python
pos_predict = np.zeros((catalog.shape[0],2))
info_predict = np.zeros(np.shape(catalog))

j = 0 # Iteration variable for positions array
for i in range(catalog.shape[0]): # Cycles through each object in catalog
    # Checks to make sure point is "good"
    # Good defined by: 1. S/N > 3     2. Sharpness < .3     3. Roundness < 1     4. Crowding < .1 
    #                  5. Object type = "Bright Star"/"Dim Star"     6. ML algorithm picks as good
    if ((catalog[i][5] >= 3)&(abs(catalog[i][6]) < .3)&(abs(catalog[i][7]) < 1)
        &(catalog[i][9] < .1)&(catalog[i][10] == 1 or catalog[i][10] == 2)): #&(cat_predictions[i] == 'green')
        pos_predict[j][0] = catalog[i][2] # Assigns X position
        pos_predict[j][1] = catalog[i][3] # Assigns Y position
        info_predict[j] = catalog[i]
        j = j + 1
                
# Trims all zeros from end of positions array, left from "bad" objects
pos_predict = pos_predict[~np.all(pos_predict == 0, axis=1)]
info_predict = info_predict[~np.all(info_predict == 0, axis=1)]

# Saves high S/N object X,Y positions
np.savetxt("New_Start(2017)/sn2008ge_Prediction.reg", pos_predict, '%5.2f')
```

```python
kwargs = {"names": ["EEP", "log10_isochrone_age_yr", "initial_mass", "log_Teff", "log_g",
                    "log_L", "z_surf", "ACS_WFC_F435W", "ACS_WFC_F475W", "ACS_WFC_F502N",
                    "ACS_WFC_F550M", "ACS_WFC_F555W", "ACS_WFC_F606W", "ACS_WFC_F625W", 
                    "ACS_WFC_F658N", "ACS_WFC_F660N", "ACS_WFC_F775W", "ACS_WFC_F814W",
                    "ACS_WFC_F850LP", "ACS_WFC_F892N", "phase"],
         "delim_whitespace": True, "comment": "#"}
isochrones = {}
for filename in glob.glob(mist_dir + "/*.iso.cmd"):
    feh_string = filename.split("/")[-1].split("_")[3] # Pulls metalicity information
    if feh_string[0] == "p":
        feh = float(feh_string[1:]) # feh is [Fe/H]
    elif feh_string[0] == "m":
        feh = -float(feh_string[1:])
    else:
        raise ValueError
    df = pandas.read_csv(filename, **kwargs)
    df['ages'] = 10 ** df.log10_isochrone_age_yr / 1e9
    isochrones[feh] = df # Creates dictionary accessible by entering a particular metalicity
```

```python
# Sampling Kroupa IMF
import scipy.integrate as integrate
def kroupa(m):
    norm = 0.32705769605379353 # integrated kroupa from .075 to 100
    if m < .08:
        alpha = .3
        return m**(-alpha)/norm
    elif m < .5:
        alpha = 1.3
        return .08 * m**(-alpha)/norm
    else:
        alpha = 2.3
        return .5 * .08 * m**(-alpha)/norm
def invertedKroupa(u):
    if u > kroupa(.08):
        return - np.log(u)/.3
    elif u > kroupa(.5):
        return - np.log(u)/(1.3 * .08)
    else:
        return - np.log(u)/(2.3 * .04)
    
m_array = np.linspace(.075,50,1000)
prob = np.zeros(1000)
for i in range(m_array.size):
    prob[i] = kroupa(m_array[i])
plt.rcParams["figure.figsize"] = [7,5]
plt.loglog(m_array, prob/1000,'g.')
plt.xlabel('Mass (in solar masses)')
plt.ylabel('Probability Density')
```

```python
masses = np.zeros(10000)
for x in range(10000):
    a = invertedKroupa(np.random.random())
    while a < .075: # Only uses stars with mass > .075 solar mass
        a = invertedKroupa(np.random.random())
    masses[x] = a
avg_mass = np.average(masses)
plt.hist(masses, bins=100, log=True)
plt.xlim(0,100)
plt.xlabel('Mass (in solar masses)')
plt.ylabel('Number of stars per 10000')
```

```python
# Calculates total luminosity and mass of galaxy
d = 17.37e6 # In units of pc
M_solar_K = 3.28 # Solar absolute magnitude in K band from http://www.ucolick.org/~cnaw/sun.html
M_solar_B = 5.45 # also from http://www.ucolick.org/~cnaw/sun.html
m_K = 7.630 # http://ned.ipac.caltech.edu/cgi-bin/datasearch?search_type=Photo_id&objid=10475&objname=NGC%201527&img_stamp=YES&hconst=73.0&omegam=0.27&omegav=0.73&corr_z=1&of=table
m_B = 11.73
M_K = m_K - 5 * np.log10(d) + 5 # Converts NGC 1527 apparent mag to absolute mag
L_K = 10**(.4*(M_solar_K - M_K)) # In units of solar luminosity
total_mass = L_K * 10**(.212*(m_B - m_K) - .959) # In units of solar mass, from Mannucci et.al. 2005
```

```python
# Calculates luminosity and mass of ROI
def K_lum(y,x):
    K_inter = K_interp(np.sqrt(x**2 + y**2))
    return d**2 * 10**((2/5)*(M_solar_K - K_inter - 5))
x0 = 23.0666/20 # Pixel measurements, in units of arcsec
y0 = -122.1982/20
r1 = 47.6/20
K_lum_ROI = integrate.dblquad(K_lum, y0-r1, y0+r1, lambda y: x0-np.sqrt(r1**2-(y-y0)**2), lambda y: x0+np.sqrt(r1**2-(y-y0)**2))[0]
mass_ROI = K_lum_ROI/L_K * total_mass # Assumes that B-K color is same in ROI and galaxy at large
```

```python
# Gives the mass which corresponds to the limiting magnitude for each isochrone and filter
df = isochrones[0]
ages = np.array(list(set(df.log10_isochrone_age_yr)))
ages.sort()

age_cmd = {}

ages = ages[(ages > 7) & (ages < 9)]
for i, age in enumerate(ages):
    idx = df.log10_isochrone_age_yr == age
    mag435 = df[idx].ACS_WFC_F435W + dist_adjust + F435W_ext + 3.610*Gal_ext
    mag555 = df[idx].ACS_WFC_F555W + dist_adjust + F555W_ext + 2.792*Gal_ext
    mag625 = df[idx].ACS_WFC_F625W + dist_adjust + F625W_ext + 2.219*Gal_ext
    mag814 = df[idx].ACS_WFC_F814W + dist_adjust + F814W_ext + 1.526*Gal_ext
    mass = df[idx].initial_mass
    age_cmd[age] = (mag435, mag555, mag625, mag814, mass)

m_min = np.zeros([100,5])
m_max = np.zeros([100,2])
j = 0
for age, (mag435, mag555, mag625, mag814, mass) in age_cmd.items():
    m435_diff = mag435.values - 27.3 # Previously, 435 limiting magnitude = 30.5
    m555_diff = mag555.values - 27.7 # Previously, 555 limiting magnitude = 30.1
    m625_diff = mag625.values - 27.5 # Previously, 625 limiting magnitude = 30.1
    m814_diff = mag814.values - 26.9 # Previously, 814 limiting magnitude = 28.9
    m_min[j,0] = age
    m_max[j,0] = age
    m_max[j,1] = np.amax(mass.values)
    if np.amin(m435_diff) < .1:
        m_min[j,1] = mass.values[np.argmin(abs(m435_diff))] # Finds mass at which F435W magnitude is closest to limiting mag
    else:
        m_min[j,1] = 100
    if np.amin(m435_diff) < .1:
        m_min[j,2] = mass.values[np.argmin(abs(m555_diff))] # Finds mass at which F555W magnitude is closest to limiting mag
    else:
        m_min[j,2] = 100
    if np.amin(m625_diff) < .1:
        m_min[j,3] = mass.values[np.argmin(abs(m625_diff))] # Finds mass at which F625W magnitude is closest to limiting mag
    else:
        m_min[j,3] = 100
    if np.amin(m814_diff) < .1:
        m_min[j,4] = mass.values[np.argmin(abs(m814_diff))] # Finds mass at which F814W magnitude is closest to limiting mag
    else:
        m_min[j,4] = 100
    j = j + 1
m_min = m_min[~np.all(m_min == 0, axis=1)]
m_min = m_min[m_min[:,0].argsort()] # Sorts m_min by mass
m_max = m_max[~np.all(m_max == 0, axis=1)]
m_max = m_max[m_max[:,0].argsort()] # Sorts m_max by mass
```

```python
n_ROI = mass_ROI/avg_mass

# Finds number of stars expected in ROI for each isochrone and filter
n_detected = np.zeros(m_min.shape)
n_detected[:,0] = m_min[:,0] # n_detected[:,0] is the age for each isochrone
for i in range(m_max.shape[0]):
    integrated = np.zeros(4)
    for j in range(4):
        if m_min[i,j+1] > m_max[i,1]:
            integrated[j] = 0
        else:
            integrated[j] = integrate.quad(kroupa, m_min[i,j+1], m_max[i,1])[0]
    n_detected[i,1:] = integrated * n_ROI
```

```python
import os
import glob
import matplotlib as mpl

mist_dir = "/Users/tktakaro/Documents/Type-Iax-HST/MIST_v1.0_HST_ACSWF"
# All of this MIST isochrone code comes from Asher

kwargs = {"names": ["EEP", "log10_isochrone_age_yr", "initial_mass", "log_Teff", "log_g",
                    "log_L", "z_surf", "ACS_WFC_F435W", "ACS_WFC_F475W", "ACS_WFC_F502N",
                    "ACS_WFC_F550M", "ACS_WFC_F555W", "ACS_WFC_F606W", "ACS_WFC_F625W", 
                    "ACS_WFC_F658N", "ACS_WFC_F660N", "ACS_WFC_F775W", "ACS_WFC_F814W",
                    "ACS_WFC_F850LP", "ACS_WFC_F892N", "phase"],
         "delim_whitespace": True, "comment": "#"}
isochrones = {}
for filename in glob.glob(mist_dir + "/*.iso.cmd"):
    feh_string = filename.split("/")[-1].split("_")[3] # Pulls metalicity information
    if feh_string[0] == "p":
        feh = float(feh_string[1:]) # feh is [Fe/H]
    elif feh_string[0] == "m":
        feh = -float(feh_string[1:])
    else:
        raise ValueError
    df = pandas.read_csv(filename, **kwargs)
    df['ages'] = 10 ** df.log10_isochrone_age_yr / 1e9
    isochrones[feh] = df # Creates dictionary accessible by entering a particular metalicity
```

```python
df = isochrones[0]
ages = np.array(list(set(df.log10_isochrone_age_yr)))
ages.sort()

age_cmd = {}

ages = ages[(ages > 7) & (ages < 7.26)]
for i, age in enumerate(ages):
    idx = df.log10_isochrone_age_yr == age
    # Three adjustments: 1. Distance: added to mag 2. MW Extinction: added to both. 3. Galactic extinction
    dist_adjust = 5 * (np.log10(17.37e6) - 1)
    F435W_ext = 0.046 # extinction in F435W in NGC 1527 from NED
    F555W_ext = 0.042 # extinction in F555W in NGC 1527 from NED
    Gal_ext = 0
    # Multiply Gal_ext by A_lambda/E(B-V) values from Schlafly, Finkbeiner 2011
    mag = df[idx].ACS_WFC_F435W + dist_adjust + F435W_ext + 3.610*Gal_ext
    color = df[idx].ACS_WFC_F435W - df[idx].ACS_WFC_F555W  + F435W_ext - F555W_ext + (3.610 - 2.792) * Gal_ext
    mass = df[idx].initial_mass
    age_cmd[age] = (color, mag, mass)

norm = mpl.colors.Normalize(vmin=ages.min(), vmax=ages.max())
sm = mpl.cm.ScalarMappable(norm=norm, cmap='viridis')
sm.set_array(ages)
plt.subplots(figsize=(15, 10))

plt.fill_between(x_array, 99*np.ones(100), 25.45*np.ones(100), color='lightgrey', zorder=0)
plt.fill_between(x_array, 99*np.ones(100), 25.95 - x_array, color='lightgrey', zorder=0)
for age, (color, mag, mass) in age_cmd.items():
    plt.plot(color, mag, alpha=0.5, color=sm.to_rgba(age), zorder=1, linewidth=4)
    #plt.fill_between(color, mag_lower, mag_upper, facecolor=sm.to_rgba(age), alpha=0.2, edgecolor='none')

x_array = np.linspace(-1, 2, 100)
    
plt.xlabel("F435W - F555W")
plt.ylabel("F555W")
plt.ylim(28,22)
plt.xlim(-1, 2)
cb = plt.colorbar(sm)
cb.set_label('$\log_{10}$ Age [yr]')
plt.draw()
plt.show()
```
