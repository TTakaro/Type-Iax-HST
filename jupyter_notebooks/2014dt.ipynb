```python
Iax_dir = "/Users/tktakaro/Documents/Type-Iax-HST" #"D:/Research/Type-Iax-HST"
mist_dir = Iax_dir + "/MIST/WFC3-UVIS"

catalog = np.loadtxt(Iax_dir+"/new_files/sn2014dt/sn2014dt")
```

```python
pos_predict = np.zeros((catalog.shape[0],2))
info_predict = np.zeros(np.shape(catalog))

j = 0 # Iteration variable for positions array
for i in range(catalog.shape[0]): # Cycles through each object in catalog
    # Checks to make sure point is "good"
    # Good defined by: 1. S/N > 3     2. Sharpness < .3    3. Roundness < 1     4. Crowding < .1 
    if ((catalog[i][5] >= 3.5)&(abs(catalog[i][6]) < .3)&(abs(catalog[i][7]) < 1)
        &(catalog[i][9] < .1)):
        pos_predict[j][0] = catalog[i][2] # Assigns X position
        pos_predict[j][1] = catalog[i][3] # Assigns Y position
        info_predict[j] = catalog[i]
        j = j + 1
                
# Trims all zeros from end of positions array, left from "bad" objects
pos_predict = pos_predict[~np.all(pos_predict == 0, axis=1)]
info_predict = info_predict[~np.all(info_predict == 0, axis=1)]

# Saves high S/N object X,Y positions
np.savetxt("Region_files/sn2014dt_Prediction.reg", pos_predict, '%5.2f')
```

```python
asci = open(Iax_dir + "/jupyter_notebooks/Region_files/sn2014dt_200pc.reg", "r+")

def mysplit(s, delim=None):
    return [x for x in s.split(delim) if x]

text = asci.readlines()
region = np.zeros([8,2])
for i in range(8):
    region[i][0] = mysplit(text[i], ' ')[0]
    region[i][1] = mysplit(text[i], ' ')[1]
region = region[~np.all(region == 0, axis=1)]

# Info for each star chosen as real 
info = np.zeros([region.shape[0],info_predict.shape[1]])

j = 0 # Iteration variable for region array
for i in range(info_predict.shape[0]): # Cycles through each star predicted above
    if (region[j][0] == pos_predict[i][0])&(region[j][1] == pos_predict[i][1]):
        info[j] = info_predict[i]
        j = j + 1
        if j == region.shape[0]: break
```

```python
# Using IRAF to compute actual concentration index, i.e. photometry at .5 pixels to photometry at 3 pixels
ascii_1 = open(Iax_dir + "/new_files/sn2014dt/final_WFC3_F438W_crc.fits1.mag.1")
ascii_2 = open(Iax_dir + "/new_files/sn2014dt/final_WFC3_F438W_crc.fits1.mag.2")
text_1 = ascii_1.readlines()
text_2 = ascii_2.readlines()

concentration = np.zeros(8)
for i in range(8):
    if mysplit(text_1[5*i + 79])[4] == 'INDEF': # Removes objects with no defined magnitude at .5 pixels
        concentration[i] = 99
    elif mysplit(text_2[5*i + 79])[4] == 'INDEF': # Removes objects with no defined magnitude at 3 pixels
        concentration[i] = -99
    else: concentration[i] = float(mysplit(text_1[5*i + 79])[4]) - float(mysplit(text_2[5*i + 79])[4])
ascii_1.close()
ascii_2.close()
```

```python
# Cut out clusters (identified as objects with concentration parameter < 2.3, from Chandar et. al. 2010)
new_info = np.zeros(info.shape)
for i in range(8):
    if (concentration[i] < 2.3): # & (concentration[i] >= 0):
        new_info[i] = info[i]
    else: print("Nope!")
new_info = new_info[~np.all(new_info == 0, axis=1)]

np.savetxt("Region_files/sn2014dt_NoClustersROI_new.reg", info[:,2:4], '%5.2f')
```

```python
df = isochrones[0.16]
ages = np.array(list(set(df.log10_isochrone_age_yr)))
ages.sort()

age_cmd = {}

ages = ages[(ages > 6.49) & (ages < 7.26)]
for i, age in enumerate(ages):
    idx = df.log10_isochrone_age_yr == age
    # Three adjustments: 1. Distance: added to mag 2. MW Extinction: added to both. 3. Galactic extinction
    dist_adjust = 5 * (np.log10(19.3) - 1)
    F275W_ext = 0.123 # extinction in F275W in M61 from NED
    F438W_ext = 0.081 # extinction in F555W in M61 from NED
    Gal_ext = 0
    # Multiply Gal_ext by A_lambda/E(B-V) values from Schlafly, Finkbeiner 2011
    mag = df[idx].WFC3_UVIS_F275W + dist_adjust + F275W_ext + 5.487*Gal_ext
    color = df[idx].WFC3_UVIS_F275W - df[idx].WFC3_UVIS_F438W  + F275W_ext - F438W_ext + (5.487 - 3.623) * Gal_ext
    mass = df[idx].initial_mass
    age_cmd[age] = (color, mag, mass)

norm = mpl.colors.Normalize(vmin=ages.min(), vmax=ages.max())
sm = mpl.cm.ScalarMappable(norm=norm, cmap='viridis')
sm.set_array(ages)
plt.subplots(figsize=(15, 10))

x_array = np.linspace(-1,2,100)
plt.fill_between(x_array, 99*np.ones(100), 24 - x_array, color='lightgrey', zorder=0)
for age, (color, mag, mass) in age_cmd.items():
    plt.plot(color, mag, alpha=0.5, color=sm.to_rgba(age), zorder=1, linewidth=4)
x_array = np.linspace(-1, 2, 100)
    
plt.xlabel("F275W - F438W")
plt.ylabel("F438W")
plt.ylim(28,22)
plt.xlim(-1, 2)
cb = plt.colorbar(sm)
cb.set_label('$\log_{10}$ Age [yr]')
plt.draw()
plt.show()
```

```python
import scipy.integrate as integrate
def kroupa(m):
    norm = 0.32705769605379353 # integrated kroupa from .075 to 100
    if m < .08:
        alpha = .3
        return m**(-alpha)/norm
    elif m < .5:
        alpha = 1.3
        return .08 * m**(-alpha)/norm
    else:
        alpha = 2.3
        return .5 * .08 * m**(-alpha)/norm
def invertedKroupa(u):
    if u > kroupa(.08):
        return - np.log(u)/.3
    elif u > kroupa(.5):
        return - np.log(u)/(1.3 * .08)
    else:
        return - np.log(u)/(2.3 * .04)
    
print(integrate.quad(kroupa, 0.075, 100))
```

```python
ages = np.array(list(set(df.log10_isochrone_age_yr)))
ages = ages[(ages > 6.49) & (ages < 8.51)]
ages.sort()
age_cmd = {}
dist_adjust = 5 * (np.log10(19.3e6) - 1)

ages_ruledout = []
for i, age in enumerate(ages):
    idx = df.log10_isochrone_age_yr == age
    Gal_ext = 0 
    mag2 = df[idx].WFC3_UVIS_F275W + dist_adjust + F275W_ext + 5.487*Gal_ext
    mag4 = df[idx].WFC3_UVIS_F438W + dist_adjust + F438W_ext + 3.623*Gal_ext
    mass = df[idx].initial_mass
    age_cmd[age] = (mag2, mag4, mass)

for age, (mag2, mag4, mass) in age_cmd.items():
    indices2 = []
    for i, mag in enumerate(mag2):
        if mag < limit_mags[0]:
            indices2.append(i)
    indices4 = []
    for i, mag in enumerate(mag4):
        if mag < limit_mags[1]:
            indices4.append(i)
            
    masses2 = []
    masses4 = []
    for i, mass in enumerate(mass):
        if np.isin(i, indices2):
            masses2.append(mass)
        if np.isin(i, indices4):
            masses4.append(mass)
            
    total_mass = 0
    n_inrange = 0
    while total_mass < 50:
        a = invertedKroupa(np.random.random())
        while a < .075 or a > 100: # Only uses stars with mass between .075 and 100 solar masses
            a = invertedKroupa(np.random.random())
        total_mass += a
        if not masses2 == []:
            if a > np.min(masses2) and a < np.max(masses2):
                n_inrange += 1
                continue
        if not masses4 == []:
            if a > np.min(masses4) and a < np.max(masses4):
                n_inrange += 1
                continue

    if n_inrange > 0:
        ages_ruledout.append(age)
```
